Los precios en pesos mexicanos del tipo de cambio del dólar americano (USD) para solventar obligaciones y publicados del 4 de enero de 2016 al 17 de enero de 2018 en los días hábiles bancarios se listan en el archivo tipoUSD.txt:

A partir de los n valores listados, calcule el valor promedio del tipo de cambio (calculando la suma y dividiendo entre la cantidad de valores) y llámelo media:

```python
arch=open("tipoUSD.txt","r") 
print(arch) 
numeros=arch.read()
arch.close 
listaNum=numeros.split(",") 
tipoUSD=[float(val) for val in listaNum] 
```

```python
media=sum(tipoUSD)/len(tipoUSD)
print(media)
print(mean(tipoUSD))
```

Para conocer la dispersión de los valores, calcule la desviación estándar, σ:

    i. Primero, calculando la lista de los cuadrados de las diferencias del valor del tipo de
    cambio y su media;
    ii. Segundo, calculando la raíz cuadrada del promedio de los cuadrados calculados en i):

```python
#inciso i
lista=[(i-media)**2 for i in tipoUSD]
```

```python
#inciso ii
desviacion=sqrt((1/len(tipoUSD))*sum(lista))
print(desviacion)
print(statistics.stdev(tipoUSD))
```

Con base en los días hábiles del periodo considerado, calcule la proporción de los días con un precio del dólar americano (USD) dentro de una banda de dos desviaciones estándar y centrada en la media:

```python
def proporcion(lista):
    c=0
    for i in tipoUSD:
        if media-desviacion<=i<=media+desviacion:
              c=c+1
    return (c/len(tipoUSD))*100
print(proporcion(tipoUSD)) 
print("La proporción de los días con un precio del dólar americano (USD) dentro de una banda de dos desviaciones estándar y centrada en la media es aproximadamente del",round(proporcion(tipoUSD)),"%")
```
