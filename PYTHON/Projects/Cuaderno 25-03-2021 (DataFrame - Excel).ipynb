50. Las cantidades dadas en el archivo de texto inflac.txt indican la tasa de inflación mensual*, es decir, una medida de la tasa de cambio mensual de los precios para el periodo de diciembre de 2005 a junio de 2018; nótese que las cantidades están ordenadas según los meses sucesivos de este periodo indicado.
Por ejemplo, el porcentaje de incremento del precio del terreno durante el mes de diciembre de 2005 fue 0.59 %, lo cual significa un incremento de valor en este mes de 0.0059; es decir, durante ese mes el terreno cambió su valor en un factor de 1.0059.

    a) Define una sucesión con las tasas dadas,
    ¿Cuáles son los valores máximo y mínimo de las tasas?
    
    b) Define una lista con las tasas ordenadas de manera descendente.
    
    c) Si un terreno estaba valuado en $60,000 (sesenta mil pesos M.N.) al final de noviembre de 2005, 
    incrementándose, o disminuyéndose, el valor en cada uno de los meses siguientes durante varios años, según 
    la tasa de cambio y hasta junio de 2018,
    ¿Qué valor tenía el terreno al final de junio de 2018 después de varios años?

```python
#inciso a
archInflac=open("inflac.txt","r") 
print(archInflac) 
datosInf=archInflac.read()
archInflac.close 
inflacionTxt=datosInf.split(",") 
inflacion=[float(i) for i in inflacionTxt] 
print("\n")
print(inflacion)

maximo=max(inflacion)
minimo=min(inflacion)

print("\n")
print("La inflación más grande en el periodo fue de: ",maximo," y la más pequeña de: ",minimo)
```

```python
#inciso b
inflacionOrdenada=inflacion.copy()
inflacionOrdenada.sort(reverse=True)
print(inflacionOrdenada)
```

```python
#inciso c
valor=60000
inf=[i/100 for i in inflacion]
n=0
while n<len(inflacion):
    valor=valor*(1+inf[n])
    n=n+1
print(valor)
```

# Ejercicio 51

Se tiene un archivo de texto llamado sueldos.txt con los sueldos que paga una empresa. Después de leer el archivo, calcule la suma de los sueldos que paga la empresa.

```python
ls
```

```python
archSueldos=open("sueldos.txt","r") 
print(archSueldos) 
datosSueldos=archSueldos.read()
archSueldos.close 
sueldosTxt=datosSueldos.split(",") 
sueldos=[float(i) for i in sueldosTxt]
print("\n")
print(sueldos)
```

```python
sumaSueldos=sum(sueldos)
print("La suma de los sueldos que paga la empresa es de: $",sumaSueldos)
```
