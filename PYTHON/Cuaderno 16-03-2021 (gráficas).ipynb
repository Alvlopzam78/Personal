```python
arch2=open("tipoUSD.txt","r") 
print(arch2) 
numeros=arch2.read()
arch2.close 
listaNum=numeros.split(",") 
tipoUSD=[float(val) for val in listaNum] 
print(tipoUSD) 
len(tipoUSD)
```

```python
plot(tipoUSD)
```

```python
#linea punteada, color rojo, marcador*
plot(tipoUSD,"-.b+")
title("Cotizaciones Dólares")
xlabel("Número de Día")
ylabel("Tipo de Cambio")
axis([0,250,17,22])
legend(["Tipo de Cambio 2017"])
```

```python
dias=list(range(1,len(tipoUSD)+1))
```
